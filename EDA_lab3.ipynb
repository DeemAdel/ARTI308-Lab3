```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
print("Shape (rows, columns):", df.shape)
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```

```python
print("New shape:", df.shape)
print("Duplicate rows:", df.duplicated().sum())
```

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['chol'], bins=20, kde=True)
plt.title("Cholesterol Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['age'])
plt.title("Boxplot of Age")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['trestbps'])
plt.title("Boxplot of Resting Blood Pressure")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='thalach', data=df)
plt.title("Age vs Maximum Heart Rate")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='chol', data=df)
plt.title("Age vs Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='target', data=df)
plt.title("Heart Disease Distribution")
plt.xticks([0,1], ['No Disease','Has Disease'])
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='target', y='chol', data=df)
plt.xticks([0,1], ['No Disease','Has Disease'])
plt.title("Cholesterol vs Heart Disease")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='target', y='thalach', data=df)
plt.xticks([0,1], ['No Disease','Has Disease'])
plt.title("Max Heart Rate vs Heart Disease")
plt.show()
```
